```python
def sort(arr):
    n=len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1,n):
            if arr[min_idx]>arr[j]:
                arr[min_idx],arr[j]=arr[j],arr[min_idx]
    print(arr)

arr=[27,54,82,12,18,79,48,10,2,90,8]
sort(arr)
```

```python
def sort(arr):
    n=len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
    print(arr)

arr=[27,54,82,12,18,79,48,10,2,90,8]
sort(arr)
```
