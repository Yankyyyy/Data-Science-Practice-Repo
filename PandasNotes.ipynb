```python
#Pandas

#Importing libraries
#python library for numerical and scientific computing. pandas is built on top of numpy
import numpy as np 
#importing pandas
import pandas as pd

series = pd.Series(data = [78, 92, 36, 64, 89])  
print("Series = \n", series)
print("Series values = \n", series.values)
print("Series index = \n", series.index)
#accessing elements using index
print("Value in series index 1 = \n", series[1])
print("Value in series index 1 to 2 = \n", series[1:3])

data = pd.Series(data = [700000, 800000, 1600000, 1800000, 30000000], index = ['Swift', 'Jazz', 'Civic', 'Altis', 'Gallardo'])
print("Series data = \n", data)
#accessing elements using user defined index
print("Value in series index Swift = \n", data['Swift'])
print("Value in series index Jazz to Gallardo = \n", data['Jazz': 'Gallardo'])
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('auto_mpg.csv')
print("Dataframe = \n", df)
print("Dataframe Head = \n", df.head())
print("Dataframe Tail = \n", df.tail())
print("Dataframe in details = \n", df.describe())
print("Dataframe information = \n", df.info())
df.dropna(inplace = True)
print("Dataframe information = \n", df.info())
print("Series containg car names = \n", df['name'])
print("Display car names = \n", df[['name']])
print("Display multiple columns = \n", df[['name', 'origin', 'model_year', 'mpg']])
```

```python
#creating a subset using head
df_head = df.head()
#Setting name as custom index
df_head.set_index('name', inplace = True)
print("Dataframe head = \n", df_head)
#accessing using iloc i.e by integer index. Syntax = df.iloc[Rows, Columns]
print("Dataframe element in row index 2 and column index 1 = \n", df.iloc[2,1])
print("Dataframe element in row index 2 and column index -1 = \n", df.iloc[2,-1])
print("Dataframe element in row index 1 to 4 and column index 4 to 5 = \n", df.iloc[1:5, 4:6])
#accessing using loc i.e by custom index
print("Dataframe element from row buick skylark 320 to amc rebel sst = \n", df_head.loc['buick skylark 320': 'amc rebel sst'])
#Subsetting from the full dataset
print("Dataframe element from row 0 to 5 and colums mentioned = \n", df.loc[0:5, ['cylinders', 'horsepower', 'name']])
#Retrieve details of all the cars built in year 72
print("All the cars built in year 72 = \n", df.loc[df['model_year'] == 72 ].head())
#Retrieve details of all the cars built in Japan having 6 cylinders
print("All the cars built in Japan having 6 cylinders = \n", df.loc[(df['origin'] == 'japan') & (df['cylinders'] == 6)].head())
#Fuel efficient
#MPG > 29, Horsepower < 93.5,
#Weight < 2500
print("All fuel efficient cars = \n", df.loc[(df['mpg'] > 29) & (df['horsepower'] < 93.5) & (df['weight'] < 2500)].head())
#Muscle cars
#Displacement >262, Horsepower > 126, Weight in range[2800, 3600]
print("All Muscle cars = \n", df.loc[(df['displacement'] > 262) & (df['horsepower'] > 126) & (df['weight'] >=2800) & (df['weight'] <= 3600)].head())
#SUV
#Horsepower > 140 , Weight > 4500
print("All SUVs = \n", df.loc[(df['horsepower'] > 140) & (df['weight'] >=4500)].head())
#Race car
#Weight <2223, acceleration > 17
print("All race cars = \n", df.loc[(df['acceleration'] > 17) & (df['weight'] < 2223)].head())
#sorted according to the number of cylinders
print("Sorted according to the number of cylinders = \n", df.sort_values(by = 'cylinders'))
#sorted in ascending order of acceleration and descending order of horsepower
print("Sorted in ascending order of acceleration and descending order of horsepower = \n", df.sort_values(['acceleration', 'horsepower'], ascending = (1,0)))
#Using list comprehension to get the numerical columns
#Syntax = DataFrame.agg(func, axis = 0)
list1 = [col for col in df.columns if df[col].dtype in ['float', 'int64']]
print("Aggregated minimum and maximum = \n", df[list1].agg(['min', 'max']))
print("Number of cars manufactured in each year = \n", df.groupby(['model_year']).count()[['name']])
#Creating a DataFrame grouped on cylinders and model_year and finding mean, min and max of horsepower
#Syntax = DataFrame.groupby(by = column_name, axis, sort)
grouped_multiple = df.groupby(['cylinders', 'model_year']).agg({'horsepower': ['mean', 'min', 'max']})
#Naming columns in grouped DataFrame
grouped_multiple.columns = ['hp_mean', 'hp_min', 'hp_max']
#Resetting index
grouped_multiple = grouped_multiple.reset_index()
#Viewing head of resulting DataFrame
print("Mean, minimum and maximum horsepower based on number of cylinders and model year = \n", grouped_multiple.head())
print("Sorted based on average acceleration of cars built in each model year = \n", df.groupby(['model_year']).mean().sort_values('acceleration', ascending = False)[['acceleration']])
#cross tab is used to gives us a tabular representation of the frequency distribution
print("Tabular representation of the frequency distribution =\n", pd.crosstab(df['model_year'], df['cylinders']))
#Creating a new dataframe with only numerical columns
num_col = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
num_col_df = df.get(num_col)
#pivot_table gives the mean of all the numerical attributes of cars for each year.
#Syntax = pd.pivot_table(data, index, aggfunc)
pivot1 = pd.pivot_table(df, index = 'model_year', aggfunc=np.mean)
print("Mean of all the numerical attributes of cars for each year =\n ", pivot1)

```

```python
marks = {'Chemistry': [67,90,66,32], 
        'Physics': [45,92,72,40],  
        'Mathematics': [50,87,81,12],  
        'English': [19,90,72,68]}
marks_df = pd.DataFrame(marks, index = ['Subodh', 'Ram', 'Abdul', 'John'])
print("Dataframe = \n", marks_df)
marks_df['Total'] = marks_df['Chemistry'] + marks_df['Physics'] + marks_df['Mathematics'] + marks_df['English']
print("Dataframe with new column Total = \n", marks_df)
marks_df.drop(columns = 'Total', inplace = True)
print("Dataframe after removing Total = \n", marks_df)
#Masking i.e where pass mark is 33 and failed marks aren't displayed
#Syntax = DataFrame.mask(cond, other = nan, inplace = False, axis = None)
f = marks_df < 33
print("Dataframe with failed marks aren't displayed = \n", marks_df.mask(f, 'Fail'))
#encrypting marks as sine of marks
encrypted_marks = np.sin(marks_df)
print("Encrypted marks = \n", encrypted_marks)
#Broadcasting
new_marks1 = marks_df + 5
print("Broadcasting Type 1 = \n", new_marks1)
new_marks2 = marks_df + [5,10,10,2]
print("Broadcasting Type 2 = \n", new_marks2)
#axis: If 0 or ‘index’: apply function to each column i.e add all the values in a column.
#If 1 or ‘columns’: apply function to each row i.e add up all the values in a row.
print("Sum of subject marks = \n", marks_df.apply(np.sum, axis = 0))
print("Sum on student marks = \n", marks_df.apply(np.sum, axis = 1))
print("Average marks based on their previous marks = \n", marks_df.apply(func = np.mean, axis = 0, result_type = 'broadcast'))
```

```python
#Concatinate multiple dataframes i.e. combines two tables. Syntax = pd.concat(data1, data2, sort)
#Case 1
marks_A = {'Chemistry': [67,90,66,32], 
        'Physics': [45,92,72,40],  
          }
marks_A_df = pd.DataFrame(marks_A, index = ['Subodh', 'Ram', 'Abdul', 'John'])
marks_B = {'Chemistry': [72,45,60,98], 
        'Physics': [78,34,72,95],  
          }
marks_B_df = pd.DataFrame(marks_B, index = ['Nandini', 'Zoya', 'Shivam', 'James'])
print("After concatination in 1st case = \n", pd.concat([marks_A_df,marks_B_df], sort = False))
#Merge multiple dataframes i.e  joins two tables based on a key. Syntax = pd.merge(data1, data2, how = 'inner')
#Case 2
df1 = pd.DataFrame({'Employee': ['Jyoti', 'Sapna', 'Raj', 'Ramaswamy'],
                    'Group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'Employee': ['Jyoti', 'Sapna', 'Raj', 'Ramaswamy'],
                    'Hire_date': [2004, 2008, 2012, 2014]})
display(df1, df2)
print("\nAfter concatination in 2nd case = \n", pd.concat([df1,df2], sort = False))
df3 = pd.merge(df1,df2)
print("\nAfter merging = \n", df3)

```
