```python
#Creating a 2D array consisting car names and horsepower
car_names = ['chevrolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino']
horsepower = [130, 165, 150, 150, 140]
car_hp_arr = np.array([car_names, horsepower])
print("Array = ", car_hp_arr)
#Accessing car names
print("Car names = ", car_hp_arr[0])
#Accessing horsepower
print("Car horsepowers = ", car_hp_arr[1])
#Accessing second car - 0 represents 1st row and 1 represents 2nd element of the row
print("Accessing elements = ", car_hp_arr[0,1])
```

```python
#Creating a 2D array consisting car names, horsepower and acceleration
cars = ['chevrolet chevelle malibu', 'buick skylark 320', 'plymouth satellite', 'amc rebel sst', 'ford torino']
horsepower = [130, 165, 150, 150, 140]
acceleration = [18, 15, 18, 16, 17]
car_hp_acc_arr = np.array([cars, horsepower, acceleration])
print("Array = \n", car_hp_acc_arr)
#accessing a subset of cars from the array
print("Type 1 = ", cars[1:4])
#Accessing name and horsepower 
print("Type 2 = ", car_hp_acc_arr[0:2])
#Accessing name and horsepower of last two cars 
print("Type 3 = ", car_hp_acc_arr[0:2, 3:5])
#Accessing name, horsepower and acceleration of first three cars
print("Type 4 = ", car_hp_acc_arr[0:3, 0:3])
```

```python
#creating a list of 5 horsepower values
horsepower = [130, 165, 150, 150, 140]
#creating a numpy array from horsepower list
horsepower_arr = np.array(horsepower)
#mean horsepower
print("Mean horsepower = ",np.mean(horsepower_arr))
#median horsepower
print("Median horsepower = ",np.median(horsepower_arr))
#finding the minimum and maximum values:
print("Minimum horsepower: ", np.min(horsepower_arr))
print("Maximum horsepower: ", np.max(horsepower_arr))
#finding the index of minimum and maximum values
print("Index of Minimum horsepower: ", np.argmin(horsepower_arr))
print("Index of Maximum horsepower: ", np.argmax(horsepower_arr))
#finding the horsepower of cars that are greater than or equal to 150
x = np.where(horsepower_arr >= 150)
print("Index of h_p greater than 150 = ", x)
# With the indices , we can find those values 
print("Filtered array = ", horsepower_arr[x])
#or creating filter array
filter_arr = horsepower_arr > 135
newarr = horsepower_arr[filter_arr]
print("Boolean of h_p greater than 135 = ", filter_arr)
print("Filtered array = ", newarr)
#Sorting without changing the original array
print('Sorted array: ', np.sort(horsepower_arr))
#Sorting while changing the original array
horsepower_arr.sort()
print('original array after sorting: ', horsepower_arr)
#Sum
print("Sum of h_p = ", np.sum(horsepower_arr))
#Addition
h_p_corrections = [5, 5, 10, 5, 10]
print("Sum of h_p = ", np.add(horsepower_arr, h_p_corrections))
```

```python
#Importing path and skimage i/o library 
import os.path
from skimage.io import imread
from skimage import data_dir
#reading the astronaut image
img = imread(os.path.join(data_dir, 'astronaut.png'))
#Using matplotlib.pyplot to visualize the image 
import matplotlib.pyplot as plt
print('Type of image: ', type(img))
print('Dimensions of image: ', img.ndim)
print('Shape of image:', img.shape)
print(img)
plt.figure()
plt.imshow(img)

#Slicing out the rocket 
img_slice = img.copy()
img_slice = img_slice[0:300,360:480]
plt.figure()
plt.imshow(img_slice)

#Image without the rocket
img[0:300,360:480,:] = 0
plt.figure()
plt.imshow(img)

#Image after putting back the rocket
img[0:300,360:480,:] = img_slice
plt.figure()
plt.imshow(img)
```

```python
#Importing Numpy
import numpy as np
#Miscellaneous methods
#start and end limit
print("Arrange type 1 = ", np.arange(0,10000))
print("Arrange type 2 = ", np.arange(0,10,2))
#Generating values between 0 and 10
arr = np.linspace(0,10)
print("Random Array with values between 1 to 10\n =", arr)
print('Length of arr: ',len(arr))
#Number of values = 3
print("3 evenly spaced numbers between 1 to 10 \n=", np.linspace(0,10,3))
print("Random numbers whose mean is 3, standard deviation is 1 and size is 100 \n =", np.random.normal(3, 1, 100))
#Returns an array of given shape filled with zeros or ones
print("Random matrix with values 0 and dimension 2x3\n =", np.zeros([2,3]))
print("Random matrix with values 1 and dimension 4x3\n =", np.ones([4,3]))
#3x3 identity matrix
print(np.eye(3))
#random integer values high=100, shape = (3,5)
x = np.random.randint(100, size=(3, 5))
print(x)
#returns a single random value from the array
x = np.random.choice([9, 3, 7, 5])
print(x)

```
